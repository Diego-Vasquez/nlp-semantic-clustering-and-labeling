```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# 1. Cargar el dataset filtrado (df_otros)
# Aquí asumimos que df_otros ya está cargado con las respuestas de la categoría 16
# La variable 't1text' contiene las respuestas abiertas.

# Asegurémonos de que 'df_otros' ya esté en el entorno de trabajo.
# En este caso, solo extraemos las respuestas que no son NaN.
respuestas_abiertas = df_otros['t1txt'].dropna()

# 2. Preprocesamiento de las respuestas abiertas (limpieza básica)


def limpiar_texto(texto):
    # Convertir a minúsculas
    texto = texto.lower()

    # Eliminar números
    texto = re.sub(r'\d+', '', texto)

    # Eliminar caracteres especiales y números (esto ya lo cubre el paso anterior)
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)

    # Eliminar puntuación (esto también se maneja con la expresión anterior)
    # (Por lo tanto, este paso es redundante y puede eliminarse)

    # Tokenizar el texto
    tokens = nltk.word_tokenize(texto)

    # Eliminar stopwords (palabras vacías)
    stop_words = set(stopwords.words('spanish'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lematización
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Reconstruir el texto limpio
    cleaned_text = ' '.join(tokens)

    # Devolver el texto limpio
    return cleaned_text


# Aplicar la limpieza a las respuestas
respuestas_limpias = respuestas_abiertas.apply(limpiar_texto)

# 3. Cargar el dataset 20 Newsgroups y preprocesarlo
newsgroups = fetch_20newsgroups(subset='all')

# Visualizar las categorías de 20 Newsgroups
categorias_newsgroups = newsgroups.target_names
print(f"Categorías de 20 Newsgroups: {categorias_newsgroups}")

# 4. Vectorizar las respuestas y los datos de 20 Newsgroups usando TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')

# Convertir las respuestas abiertas en vectores TF-IDF
X_respuestas = vectorizer.fit_transform(respuestas_limpias)

# Convertir el dataset 20 Newsgroups en vectores TF-IDF
X_newsgroups = vectorizer.transform(newsgroups.data)

# 5. Entrenar un modelo SVM
y_newsgroups = newsgroups.target  # Etiquetas de 20 Newsgroups

# Dividir el dataset de 20 Newsgroups en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_newsgroups, y_newsgroups, test_size=0.2, random_state=42)

# Entrenar el clasificador SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Evaluar el modelo en el conjunto de prueba
y_pred = svm_model.predict(X_test)
print("Evaluación del modelo SVM en 20 Newsgroups:")
print(classification_report(y_test, y_pred))

# 6. Definir la función para asignar "Otros" cuando la predicción no es relevante
def asignar_categoria(prediccion, categorias_newsgroups):
    # Definir un conjunto de categorías relevantes para tu encuesta, mapeando categorías de Newsgroups
    categorias_relevantes = [
        "rec.autos",  # Relacionado con la falta de empleo, temas laborales
        "sci.med",    # Relacionado con la falta de cobertura en salud pública
        "talk.politics.misc",  # Relacionado con el mal funcionamiento de la democracia
        "soc.religion.christian",  # Puede relacionarse con la corrupción, moralidad, etc.
        "sci.space",  # Relacionado con la ciencia, podría asociarse a falta de apoyo a la ciencia y tecnología
        "comp.graphics",  # Podría relacionarse con mala calidad en la educación estatal
        "misc.forsale",  # Relacionado con pobreza, aumento de precios, bajos sueldos
    ]

    # Si la categoría predicha no es relevante, asignar "Otros"
    if categorias_newsgroups[prediccion] not in categorias_relevantes:
        return "Otros"
    else:
        # Mapeo de categorías de Newsgroups a las categorías de tu encuesta
        mapeo = {
            "sci.med": "Falta de cobertura / mala atención en salud pública",  # Relacionado con la salud pública
            "rec.autos": "La falta de empleo",  # Relacionado con la falta de empleo
            "talk.politics.misc": "Mal funcionamiento de la democracia",  # Relacionado con la democracia
            "soc.religion.christian": "La corrupción",  # Relacionado con temas de corrupción o valores éticos
            "sci.space": "Falta de apoyo a la agricultura",  # Puede asociarse con temas de ciencia o agricultura
            "comp.graphics": "Mala calidad de la educación estatal",  # Relacionado con educación y tecnología
            "misc.forsale": "Bajos sueldos / aumento de precios",  # Relacionado con temas económicos y pobreza
        }

        # Devolver la categoría mapeada
        return mapeo.get(categorias_newsgroups[prediccion], "Otros")

# 7. Predicción para las respuestas abiertas y asignar categorías
predicciones = svm_model.predict(X_respuestas)

# Mapear las predicciones de categorías de Newsgroups
categorias_predichas = [categorias_newsgroups[pred] for pred in predicciones]

# Asignar "Otros" para las predicciones irrelevantes
categorias_finales = [asignar_categoria(pred, categorias_newsgroups) for pred in predicciones]

# 8. Mostrar los resultados finales
resultados = pd.DataFrame({
    'Respuesta': respuestas_abiertas,
    'Categoría Predicha': categorias_finales
})

# Ver los primeros resultados
print(resultados.head())

```
